```python
import pandas as pd

df = pd.read_excel('Downloads/NBA Data/NBA_Salaries_Final.xlsx') # can also index sheet by name or fetch all sheets
df.drop(['PLAYER_IDX','NAME'],axis=1,inplace=True)
```

```python
df['PPG'] = df['PTS']/df['GP']
```

```python
df['RPG'] = df['REB']/df['GP']
df['APG'] = df['AST']/df['GP']
df['FPPG'] = df['FP']/df['GP']
df['FGAPG'] = df['FGA']/df['GP']
df['3PMPG'] = df['3PM']/df['GP']
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Import linear regression models
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV

# Import logistic regression models
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV

# Import confusion matrix function from sklearn
from sklearn.metrics import confusion_matrix
```

```python
# Split the data into training and testing sets
df_train = df[:172]
df_test = df[172:]

# Define features to remove from the feature data (i.e., X)
drop_for_X = ['SALARY','TEAM','AGE','GP','MIN','TOV','PLAYER']

# Partition the training data into features and target
X_train = df_train.drop(columns=drop_for_X)
y_train = df_train['SALARY']

# Fit the model
linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Partition the testing data into features and target
X_test = df_test.drop(columns=drop_for_X)
y_test = df_test['SALARY']

# Predict the number of wins
y_test_predictions = linreg.predict(X_test)
```

```python
train_score = linreg.score(X_train, y_train)
test_score = linreg.score(X_test, y_test)
print("Train Score: {}\nTest Score: {}".format(train_score,test_score))
```

```python
df_train = df[:172]
df_test = df[172:]

# Partition the training data into features and target
X_train = df_train.drop(columns=['SALARY','TEAM','PLAYER'])
y_train = df_train['SALARY']

# Fit the model
lasso = LassoCV()
lasso.fit(X_train, y_train)

# Partition the testing data into features and target
X_test = df_test.drop(columns=['SALARY','TEAM','PLAYER'])
y_test = df_test['SALARY']

# calculate the score
train_score = lasso.score(X_train, y_train)
test_score = lasso.score(X_test, y_test)
print(f'The train score is {train_score:.3} and the test score is {test_score:.3}')
```
